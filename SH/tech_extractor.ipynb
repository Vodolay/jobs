```python
leaf_list = []

def list_of_dict_iterator(list_of_dict):
    for dicti in list_of_dict:
        if 'children' in dicti:
            if dicti['children']:
                list_of_dict_iterator(dicti['children'])
            else:
                leaf_list.append(dicti)
        else:
            leaf_list.append(dicti)
```

```python
list_of_dict_iterator(listy)
```

```python
len(leaf_list)
```

```python
hy = pd.DataFrame.from_records(leaf_list)
```

```python
hy = hy.drop(columns = ['children'])
```

```python
hy = hy.drop_duplicates()
```

```python
fl = pd.read_csv('full_tech_list.csv')
```

```python
fl
```

```python
value = 100000
nodeid = []
```

```python
for tech in fl['name']:
    nodeid.append(value)
    value += 1
```

```python
fl['nodeid'] = nodeid
```

```python
def get_nodeid(tech):
    key = fl[fl['name'] == tech].index.values
    entity = key[0]
    return fl.at[entity, 'nodeid']

def replace_nodeid(list_of_techs):
    global fl
    if len(list_of_techs) == 2:
        fl = fl.replace(get_nodeid(list_of_techs[1]),get_nodeid(list_of_techs[0]))
    elif len(list_of_techs) == 3:
        fl = fl.replace(get_nodeid(list_of_techs[1]),get_nodeid(list_of_techs[0]))
        fl = fl.replace(get_nodeid(list_of_techs[2]),get_nodeid(list_of_techs[0]))
    elif len(list_of_techs) == 4:
        fl = fl.replace(get_nodeid(list_of_techs[1]),get_nodeid(list_of_techs[0]))
        fl = fl.replace(get_nodeid(list_of_techs[2]),get_nodeid(list_of_techs[0]))
        fl = fl.replace(get_nodeid(list_of_techs[3]),get_nodeid(list_of_techs[0]))
```
